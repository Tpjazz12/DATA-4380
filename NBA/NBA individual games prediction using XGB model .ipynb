Create the new columns "Average 30 games of plus and minus" to calculate the average point from past 30 games by the PLUS_MINUS columns

```python
#The average plus minus scores for the pass 30 games
games['avg_30_plus_minus']=games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x: x.rolling(30,closed='left').mean())
```

The possibilities of Winning or Losing based on home or away games

```python
#Series of game away and home game
msk = games['MATCHUP'].str.contains('@')
games_away = games[msk]
games_home = games[~msk]
```

```python
games_away.shape
```

```python
games_home.shape
```

```python
games_home
```

```python
games_away
```

```python
games_merged = pd.merge(games_home,games_away,on='GAME_ID',suffixes=('_home','_away'))
games_merged
```

```python
games_merged['avg_30_plus_minus_diff'] = games_merged['avg_30_plus_minus_home']-games_merged['avg_30_plus_minus_away']
```

```python
games_merged
```

```python
games_model = games_merged[['WL_home','avg_30_plus_minus_diff']].dropna()
#compared WL_home and avg_30_plus_minus_diff and drop NA
```

```python
games_model
```

```python
games_model['WL_home'] = games_model['WL_home'].map({'W':1,'L':0})
```

```python
games_model
```

# Building a predictive model

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test = train_test_split(games_model,stratify=games_model['WL_home'],test_size=0.2,random_state=7)
```

```python
df_train.shape
```

```python
df_test.shape
```

Predict by WL_home

```python
target = 'WL_home'
X_train = df_train.drop(columns=target) #avg_30_plus_minus_diff
y_train = df_train[target] #WL_home

X_test = df_test.drop(columns=target) #avg_30_plus_minus_diff
y_test = df_test[target] #WL_home
```

```python
pip install xgboost
```

```python
import xgboost as xgb
```

```python
classify = xgb.XGBClassifier(use_label_encoder=False, random_state=7)
```

```python
classify.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = classify.predict(X_test)
```

```python
accuracy_score(y_test,y_pred) #testing the accuracy of the prediction
```

Hyperparameter tuning - Random search 

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.utils.fixes import loguniform
```

```python
hyp_params = {'learning_rate': loguniform(0.0001, 1),
              'max_depth': [2, 3, 4, 5, 6, 7, 8, 9],
              'subsample': [0.7, 0.8, 0.9, 1.0],
              'n_estimators': [50, 100, 150, 200]}
```

```python
random_hyp = RandomizedSearchCV(estimator=classify, 
                                param_distributions=hyp_params, 
                                n_iter=20, 
                                cv=7,
                                scoring='accuracy',
                                random_state=7)
```

```python
random_hyp.fit(X_train, y_train)
```

```python
random_hyp.best_params_
```

```python
model_hyp = random_hyp.best_estimator_
```

```python
y_pred_hyp = model_hyp.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_hyp)
```

Take two teams name (Home and Away) then return the prediction result of which team will win the game 

```python
#home team and away team -> result 
team_home = 'Dallas Mavericks'
team_away = 'Los Angeles Lakers'
```

```python
def predict_games(team_home,team_away):
    from nba_api.stats.endpoints import leaguegamefinder

    gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable='01/01/2022')
    games = gamefinder.get_data_frames()[0]

    games = games[['TEAM_NAME', 'GAME_ID','GAME_DATE', 'MATCHUP', 'WL', 'PLUS_MINUS']]

    import pandas as pd

    games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])

    msk_home = (games['TEAM_NAME'] == team_home)
    games_30_home = games[msk_home].sort_values('GAME_DATE').tail(30)
    home_plus_minus = games_30_home['PLUS_MINUS'].mean()

    msk_away = (games['TEAM_NAME'] == team_away)
    games_30_away = games[msk_away].sort_values('GAME_DATE').tail(30)
    away_plus_minus = games_30_away['PLUS_MINUS'].mean()

    games_different = home_plus_minus - away_plus_minus

    model_saved = load('model_nba.joblib')

    import numpy as np 
    predict_home_win = model_saved.predict(np.array([games_different]))[0] 

    predict_wining_probability = model_saved.predict_proba(np.array([games_different]))[0][1]

    return predict_home_win, predict_wining_probability
```

Predict "Dallas Mavericks"-home team vs "Los Angeles Lakers"-away team on Sunday 02/26/2023

```python
predict_games('Dallas Mavericks','Los Angeles Lakers')
```
